```python
train_tagged = open("en_train_tagged.txt", "r")
word_list = []
word_count = 0
tokens = set()
for line in train_tagged:
    for word in line.split(" "):
        word_list.append(word.split("/")[1])
        #word_list.append(" ")
        tokens.add(word)
        word_count+=1;
    word_list.append("\n")
file_content = ''.join(word_list)
print len(tokens)
print file_content
```
